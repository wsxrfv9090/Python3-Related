```python
query ='''
select a.customerid,
       count(distinct c.orderid) as orders,
       sum(c.quantity*d.price) as spend,
       sum(c.quantity*d.price)/count(distinct c.orderid) as avg_spend
from customers a
join orders b on a.customerid=b.customerid
join orderdetails c on b.orderid=c.orderid
join products d on c.productid=d.productid
group by 1
order by 1;
'''
pd.read_sql_query(query,conn)
```

```python
query = '''
select a.customerid,
       avg(d.price*c.quantity) as avg_spend
from customers a
join orders b on a.customerid=b.customerid
join orderdetails c on b.orderid=c.orderid
join products d on c.productid=d.productid
group by 1
order by a.customerid,b.orderid
'''

# run script
pd.read_sql_query(query,conn)
```
